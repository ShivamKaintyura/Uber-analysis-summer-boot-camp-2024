## Data
The data contains the details for the Uber rides across various boroughs (subdivisions) of New York City at an hourly level and attributes associated with weather conditions at that time.
### Data description
 * pickup_dt: Date and time of the pick-up.
 * borough: NYC's borough.
 * pickups: Numberof pickups for the period (hourly).
 * spd: Wind speed in miles/hour.
 * vsb: Visibility in miles to the nearest tenth.
 * temp: Temperature in Fahrenheit.
 * dewp: Dew point in Fahrenheit.
 * slp: Sea level pressure.
 * pcp01: 1-hour liquid precipitation.
 * pcp06: 6-hour liquid precipitation.
 * pcp24: 24-hour liquid precipitation.
 * sd: Snow depth in inches.
 * hday: Being a holiday (Y) or not (N).

### Q4-Hourly Trends

```python
print('the peak hours for Uber pickups in each borough')
peak_hours_by_borough =data.groupby(['borough', 'Hours'])['pickups'].sum().reset_index()
peak_hours_by_borough = peak_hours_by_borough.loc[peak_hours_by_borough.groupby('borough')['pickups'].idxmax()]
print(peak_hours_by_borough)
```

```python
print('the number of pickups change during rush hours (e.g., 7-9 AM, 5-7 PM)')
rush_hours = [7, 8, 9, 17, 18, 19]
pickups_rush_hours = data[data['Hours'].isin(rush_hours)].groupby('Hours')['pickups'].sum()
print(pickups_rush_hours)
```

```python
print('the average number of pickups during late-night hours (e.g., 12 AM - 4 AM)')
late_night_hours = [0, 1, 2, 3, 4]
pickups_late_night =data[data['Hours'].isin(late_night_hours)].groupby('Hours')['pickups'].mean()
print(pickups_late_night)
```

### Q5-Borough Comparison

```python
print('pickup trends differ between boroughs during different weather conditions')
data['temp_range'] = pd.cut(data['temp'], bins=[-float('inf'), 32, 50, 68, 86, float('inf')], labels=['Very Cold', 'Cold', 'Mild', 'Warm', 'Hot'])
pickuptrends = data.groupby(['borough', 'temp_range'])['pickups'].mean().unstack()
print(pickuptrends)
```

```python
print('borough shows the highest increase in pickups during holidays')
holiday_increase = data.groupby(['borough', 'hday'])['pickups'].mean().unstack()
holiday_increase['increase'] = holiday_increase['Y'] - holiday_increase['N']
highest_increase_borough = holiday_increase['increase'].idxmax()
print(highest_increase_borough)
```

```python
print(' the number of pickups compare between weekdays and weekends for each borough')
data['day_type'] = data['pickup_dt'].dt.dayofweek.apply(lambda x: 'Weekend' if x >= 5 else 'Weekday')
d_ww=data.groupby('day_type')["pickups"].mean().reset_index()
d_ww
```

### Q6-Weather Extremes

```python
print('extreme weather conditions (e.g., very high or very low temperatures, heavy rainfall, snowstorms) affect the number of pickups')
# extreme_weather_pickups = data[(data['temp'] < 20) | (data['temp'] > 90) | (data['pcp01'] > 1) | (data['sd'] > 5)]
# extreme_weather_pickups.groupby(["temp",'pcp01',"sd"])["pickups"].mean()
extreme_weather_pickups = data[
    (data['temp'] < 20) |  # Very low temperature
    (data['temp'] > 90) |  # Very high temperature
    (data['pcp01'] > 1) |  # Heavy rainfall
    (data['sd'] > 5)       # Snowstorms
]
# Group by temperature ranges
extreme_temp_pickups = extreme_weather_pickups[(data['temp'] < 20) | (data['temp'] > 90)]
extreme_temp_avg_pickups = extreme_temp_pickups.groupby('temp')['pickups'].mean()

# Group by heavy rainfall
heavy_rain_pickups = extreme_weather_pickups[data['pcp01'] > 1]
heavy_rain_avg_pickups = heavy_rain_pickups.groupby('pcp01')['pickups'].mean()

# Group by snowstorms
snowstorm_pickups = extreme_weather_pickups[data['sd'] > 5]
snowstorm_avg_pickups = snowstorm_pickups.groupby('sd')['pickups'].mean()
```
